```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)

m = 100 # No. oF samples
X = np.random.randn(m,1) # X column vector
Y = (4 + 3*X) + np.random.randn(m,1) # adding noise on eqn 3x + 4
```

```python
plt.scatter(X,Y)
```

```python
from sklearn.preprocessing import add_dummy_feature

X_b = add_dummy_feature(X,value = 1)
theta_best = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ Y
```
